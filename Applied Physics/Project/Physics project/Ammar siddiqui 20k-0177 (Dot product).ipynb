# DOT PRODUCT OF VECTORS:

*The dot or scalar product is defined as the Sum of the magnitudes of the coordinates of two vectors in a given dimension*

###### **Mathematically**
If there are two vectors say A and B then their dot product is written as:

$A.B= ABcos\theta$ \
Below there are two matrices taken and they are put into dot product and the result is shown in the graph.

**Note**: For Dot product the no of columns of first matrix must be equal to number of rows of the other matrix
