# **SIMPLE HARMONIC MOTION**

The simple harmonic motion is defined as: 
            *When a body repeats its motion in a particular interval of time and the acceleration produced is directly proportional to the mean position then the body is said to be in SHM*

#### **FORMULAS**:
* Displacement $x(t)$ = $Acos(2πft)$ (as a function of time)
* Velocity $dx/dt$ OR $v(t)$ = $-Asin(2πft)$ (as  a function of time)
* Acceleration $d^2x/dt^2$ OR $a(t)$ = $-Acos(2πft)$ (as a function of time)

#### **GRAPHICAL REPRESENTATION**:
The Displacement, velocity, acceleration can easily be understood with the help of a graph.
Their graph is same as of Sine and Cosine function.
These graphs are below\
Here we will show you how to calculate displacement using the formula stated above
###### Displacement vs Time graph:

```python
x= np.linspace(0,2*np.pi,1000)
a=int(input("Enter amplitude: "))
def f(x): return a*np.cos(x)
plt.plot(x,f(x))
plt.title('Displacement vs Time graph')
plt.xlabel('Time (t)')
plt.ylabel('Displacement(x)')
plt.axhline(color='black')
plt.axvline(color='black')
t=int(input("Enter time period: "))
phase=int(input("Enter phase angle:(in radians) "))
f=float(input("Enter frequency: "))
d=a*np.cos(2*(np.pi*f*t)+phase)
print("Displacement is = ",d)
```

###### Velocity vs Time graph: 

```python
a=float(input("Enter amplitude: "))
def f(x): return a*-np.sin(x)
plt.plot(x,f(x),color ='red')
plt.title('Velocity vs Time graph')
plt.xlabel('Time (t)')
plt.ylabel('Velocity (V)')
plt.axhline(color='black')
plt.axvline(color='black')
```

###### Acceleration vs Time graph

```python
a=float(input("Enter amplitude: "))
def f(x): return -(a)*(np.cos(x))
plt.plot(x,f(x),color='purple')
plt.axhline(color='black')
plt.axvline(color='black')
plt.xlabel('Time (t)')
plt.ylabel('Acceleration (a)')
plt.title('Acceleration vs Time Graph')
plt.figure(figsize=(600,600))
```
